```python
fl = 6.75; fh = 58.75
idx_f = np.logical_and(f>=fl, f<=fh);
f_idx = f[idx_f]
idx_f1_fit = np.logical_and(f_idx>=12.50, f_idx<=15.00);
idx_f2_fit = np.logical_and(f_idx>=47.50, f_idx<=57.50);
idx_f_fit = np.logical_or(idx_f1_fit, idx_f2_fit);
f_fit = f_idx[idx_f_fit]


rats_group = ['_pre','_post',]
for rats in rats_group:
    erfdn = np.load('Rats/Rats_ephy'+rats+'.npy')#[[3]];
    print('\n', rats)
    rr = 0
    for rfdn in erfdn:

        epn    = np.load(rfdn+'tracking/epn_ephy'+rats+'.npy')
        subset = np.load(rfdn+'tracking/subset_ephy_li.npy');

        n_k = len(epn)
        psd_m_l_fit = np.zeros([n_k, len(f_idx)]);
        psd_m_i_fit = np.zeros([n_k, len(f_idx)]);
        
        y_m_l_fit = np.zeros([n_k, len(f_idx)]);
        y_m_i_fit = np.zeros([n_k, len(f_idx)]);
        
        rc = int(np.sqrt(n_k))+1
        jj = 0
        psd_m_l = np.load(rfdn+'tracking'+'/psd_m_l'+rats+'.npy');
        psd_m_i = np.load(rfdn+'tracking'+'/psd_m_i'+rats+'.npy');
        
        psd_m_m_i = np.mean(psd_m_i,axis=0)[idx_f]; psd_m_s_i = np.std (psd_m_i,axis=0)[idx_f]
        psd_m_m_l = np.mean(psd_m_l,axis=0)[idx_f]; psd_m_s_l = np.std (psd_m_l,axis=0)[idx_f]

        for k in np.arange(n_k):
            fdn = rfdn + epn[k] + '/'

            regr = linear_model.LinearRegression()  
            regr.fit(np.log10(f_fit).reshape(-1,1), np.log10((psd_m_l[k,:][idx_f])[idx_f_fit]).T)
            slope_l = (regr.coef_).squeeze();       intercept_l = (regr.intercept_).squeeze()
            y_l = np.power(10, intercept_l)*np.power(f_idx, slope_l)
            
            regr = linear_model.LinearRegression()  
            regr.fit(np.log10(f_fit).reshape(-1,1), np.log10((psd_m_i[k,:][idx_f])[idx_f_fit]).T)
            slope_i = (regr.coef_).squeeze();       intercept_i = (regr.intercept_).squeeze()
            y_i = np.power(10, intercept_i)*np.power(f_idx, slope_i)
            
#             np.save(fdn+'tracking/slope_l.npy',slope_l); np.save(fdn+'tracking/intercept_l.npy',intercept_l);
#             np.save(fdn+'tracking/slope_i.npy',slope_i); np.save(fdn+'tracking/intercept_i.npy',intercept_i);

            psd_m_l_fit[k,:] = (psd_m_l[k,:][idx_f])/y_l;   y_m_l_fit[k,:] = y_l
            psd_m_i_fit[k,:] = (psd_m_i[k,:][idx_f])/y_i;   y_m_i_fit[k,:] = y_i

        np.save(rfdn+'tracking/psd_m_l_fit'+rats+'.npy',psd_m_l_fit);
        np.save(rfdn+'tracking/psd_m_i_fit'+rats+'.npy',psd_m_i_fit);
        np.save(rfdn+'tracking/y_m_l_fit'+rats+'.npy',y_m_l_fit);
        np.save(rfdn+'tracking/y_m_i_fit'+rats+'.npy',y_m_i_fit);
        print(rfdn, n_k)
        rr += 1
```

```python
fig = plt.figure(1, figsize=(1,1))
fig.set_figheight(8); fig.set_figwidth(10)
gs = GridSpec(2, 2);
periods = ['within 7 days post surgery', 'after 10 days post surgery']

group = 0
example = 3

for rats in rats_group:
    erfdn = np.load('Rats/Rats_ephy'+rats+'.npy')[[1]];
    n_r = len(erfdn)
    r = int(n_r/2)+1;   print(r)

    rr = 0
    for rfdn in erfdn:
        epn   = np.load(rfdn+'tracking'+'/epn_ephy'+rats+'.npy')
        psd_m_l = np.load(rfdn+'tracking/psd_m_l'+rats+'.npy');
        psd_m_i = np.load(rfdn+'tracking/psd_m_i'+rats+'.npy');
        psd_m_m_l = np.mean(psd_m_l,axis=0)[idx_f];  psd_m_s_l = np.std (psd_m_l,axis=0)[idx_f]
        psd_m_m_i = np.mean(psd_m_i,axis=0)[idx_f];  psd_m_s_i = np.std (psd_m_i,axis=0)[idx_f]
        
        psd_m_l_fit = np.load(rfdn+'tracking/psd_m_l_fit'+rats+'.npy');
        psd_m_i_fit = np.load(rfdn+'tracking/psd_m_i_fit'+rats+'.npy');
        psd_m_m_l_fit = np.mean(psd_m_l_fit,axis=0); psd_m_s_l_fit = np.std (psd_m_l_fit,axis=0)
        psd_m_m_i_fit = np.mean(psd_m_i_fit,axis=0); psd_m_s_i_fit = np.std (psd_m_i_fit,axis=0)

        y_m_l_fit = np.load(rfdn+'tracking'+'/y_m_l_fit'+rats+'.npy');
        y_m_i_fit = np.load(rfdn+'tracking'+'/y_m_i_fit'+rats+'.npy');
        y_m_m_l_fit = np.mean(y_m_l_fit,axis=0); y_m_s_l_fit = np.std (y_m_l_fit,axis=0)
        y_m_m_i_fit = np.mean(y_m_i_fit,axis=0); y_m_s_i_fit = np.std (y_m_i_fit,axis=0)

        ax_idx = gs[rr,group];ax = fig.add_subplot(ax_idx);
        ax.spines['right'].set_visible(False); ax.spines['top'].set_visible(False)
#         ax.set_yticks([]);
        ax.plot(f_idx, psd_m_l[example,idx_f], 'b', lw = 2);
        ax.plot(f_idx, y_m_l_fit[example,:], 'b--', lw = 2);  ax.plot(f_fit, psd_m_l[example, idx_f][idx_f_fit], 'b+', ms =10);
#         ax.plot(f_fit, y_m_l_fit[example, idx_f_fit], 'bx');
        ax.plot(f_idx, psd_m_i[example,idx_f], 'r', lw = 2);
        ax.plot(f_idx, y_m_i_fit[example,:], 'r--', lw = 2);  ax.plot(f_fit, psd_m_i[example, idx_f][idx_f_fit], 'r+', ms =10);
        ax.set_ylim([8, 40]);
        ax.set_title(periods[group], fontsize = 15);

        
        ax2 = inset_axes(ax, width="55%", height="55%",
#                   bbox_to_anchor=(30, 30, 10, 10),
                    bbox_to_anchor=(.5, .5, .8, .8),
                   bbox_transform=ax.transAxes,
                            loc=3)
        ax2.spines['right'].set_visible(False); ax2.spines['top'].set_visible(False)
        ax2.loglog(f_idx, psd_m_l[example,idx_f], 'b', basex=10, lw = 1);
        ax2.loglog(f_idx, y_m_l_fit[example,:], 'b--',   basex=10, lw = 1);
        ax2.loglog(f_idx, psd_m_i[example,idx_f], 'r', basex=10, lw = 1);
        ax2.loglog(f_idx, y_m_i_fit[example,:], 'r--',   basex=10, lw = 1);
        ax2.yaxis.set_major_formatter(NullFormatter()); ax2.yaxis.set_minor_formatter(NullFormatter())
        ax2.xaxis.set_major_formatter(NullFormatter()); ax2.xaxis.set_minor_formatter(NullFormatter())
        
        ax_idx = gs[rr+1, group];ax = fig.add_subplot(ax_idx);
        ax.spines['right'].set_visible(False); ax.spines['top'].set_visible(False)
        ax.plot(f_idx, psd_m_m_l_fit, 'b-', lw = 2);
        ax.fill_between(f_idx, psd_m_m_l_fit+psd_m_s_l_fit, psd_m_m_l_fit-psd_m_s_l_fit, color='blue', alpha='0.5')
        ax.plot(f_idx, psd_m_m_i_fit, 'r-', lw = 2);
        ax.fill_between(f_idx, psd_m_m_i_fit+psd_m_s_i_fit, psd_m_m_i_fit-psd_m_s_i_fit, color='red',  alpha='0.5')
        ax.set_xlabel('Frequency (Hz)', fontsize = 12)
        
        ax.set_ylim([0.8, 2.0]);
#         ax.set_xlabel('time: sec', fontsize = 8);  ax.set_ylabel('speed: cm/s', fontsize = 8)
        rr += 2
    group = group+1

    
ax_idx = gs[0, 0];ax = fig.add_subplot(ax_idx);
ax.set_ylabel('single session \n \n Squared Norm Power:au', fontsize = 12)
Lesion   = Line2D([0], [0], color='b', ls = '-', lw=2, label='Lesion')
Intact   = Line2D([0], [0], color='r', ls = '-', lw=2, label='Intact')
Baseline = Line2D([0], [0], color='k', ls = '--', lw=2, label='Baseline')
fit      = Line2D([0], [0], color='w', ls = '--', lw=2, label='Baseline')

ax.legend(handles=[Lesion, Intact, Baseline], loc='lower left', fontsize = 8)

ax_idx = gs[1, 0];ax = fig.add_subplot(ax_idx);
ax.set_ylabel('Average \n \n z-scored power:au', fontsize = 12)
```
