```python
rats = '_pre'
# rats = '_post'
print(rats)
erfdn = np.load('Rats/Rats_ephy'+rats+'.npy')#[[3]];
fl = 3; fh = 70

rr = 0
for rfdn in erfdn:
    epn   = np.load(rfdn+'tracking'+'/epn_ephy'+rats+'.npy')
    subset = np.load(rfdn+'tracking'+'/subset_ephy_li.npy');
    print(rfdn, len(epn))
    n_k = len(epn)
    jj = 0
    psd_m_l = np.zeros([n_k, len(f)]);
    psd_m_i = np.zeros([n_k, len(f)]);
    for k in epn:
        fdn = rfdn + k + '/'
        idx_clean_f     = np.load(fdn+'tracking/'+'idx_clean_f.npy')
        idx_clean_fit = np.load(fdn+'tracking/'+'idx_clean_fft.npy')
        idx_clean   = np.logical_and(idx_clean_f, idx_clean_fit)
        r = np.sum(idx_clean)/len(idx_clean);

        i = subset[0]; chn = str(i//10)+str(i%10)
        Y_f = np.load(fdn + 'FFT_'+str(fs_out)+'/ch.'+chn+'.fps.npy').T
        psd = np.mean(Y_f[:,idx_clean], axis=1);
        psd_m_l[jj,:] = psd
        
        i = subset[1]; chn = str(i//10)+str(i%10)
        Y_f = np.load(fdn + 'FFT_'+str(fs_out)+'/ch.'+chn+'.fps.npy').T
        psd = np.mean(Y_f[:,idx_clean], axis=1);
        psd_m_i[jj,:] = psd
        jj = jj+1

    np.save(rfdn+'tracking'+'/psd_m_l'+rats+'.npy', psd_m_l);
    np.save(rfdn+'tracking'+'/psd_m_i'+rats+'.npy', psd_m_i);
    rr += 1
```

```python
fl = 6.75; fh = 58.75
idx_f = np.logical_and(f>=fl, f<=fh);
f_idx = f[idx_f]
idx_f1_fit = np.logical_and(f_idx>=13.75, f_idx<=16.25);
idx_f2_fit = np.logical_and(f_idx>=47.50, f_idx<=57.50);
idx_f_fit = np.logical_or(idx_f1_fit, idx_f2_fit);
f_fit = f_idx[idx_f_fit]


rats_group = ['_pre','_post',]
for rats in rats_group:
    erfdn = np.load('Rats/Rats_ephy'+rats+'.npy')#[[3]];
    print('\n', rats)
    rr = 0
    for rfdn in erfdn:

        epn   = np.load(rfdn+'tracking'+'/epn_ephy'+rats+'.npy')
        subset = np.load(rfdn+'tracking'+'/subset_ephy_li.npy');

        n_k = len(epn)
        psd_m_l_fit = np.zeros([n_k, len(f_idx)]);
        psd_m_i_fit = np.zeros([n_k, len(f_idx)]);
        
        y_m_l_fit = np.zeros([n_k, len(f_idx)]);
        y_m_i_fit = np.zeros([n_k, len(f_idx)]);
        
        rc = int(np.sqrt(n_k))+1
        jj = 0
        psd_m_l = np.load(rfdn+'tracking'+'/psd_m_l'+rats+'.npy');
        psd_m_i = np.load(rfdn+'tracking'+'/psd_m_i'+rats+'.npy');
        psd_m_m_i = np.mean(psd_m_i,axis=0)[idx_f]
        psd_m_s_i = np.std (psd_m_i,axis=0)[idx_f]
        psd_m_m_l = np.mean(psd_m_l,axis=0)[idx_f]
        psd_m_s_l = np.std (psd_m_l,axis=0)[idx_f]

        d_psd_m = psd_m_l-psd_m_i;
        d_psd_m_m = np.mean(d_psd_m,axis=0)[idx_f]
        d_psd_m_s = np.std (d_psd_m,axis=0)[idx_f]

        epn_clean = np.zeros((n_k,))>-1
        fig = plt.figure(1, figsize=(rc*5,rc*5))
        gs = GridSpec(rc, rc);

        for k in np.arange(n_k):
            fdn = rfdn + epn[k] + '/'
            ax_idx = gs[k];ax = fig.add_subplot(ax_idx);ax.set_title(rfdn);
            
            ax.plot(f[idx_f], psd_m_l[k,idx_f], 'b');
            ax.fill_between(f[idx_f], psd_m_m_l+psd_m_s_l, psd_m_m_l-psd_m_s_l, color='blue', alpha='0.5')
            ax.plot(f[idx_f], psd_m_i[k,idx_f], 'r');
            ax.fill_between(f[idx_f], psd_m_m_i+psd_m_s_i, psd_m_m_i-psd_m_s_i, color='red',  alpha='0.5')

            regr = linear_model.LinearRegression()  
            regr.fit(np.log10(f_fit).reshape(-1,1), np.log10((psd_m_l[k,:][idx_f])[idx_f_fit]).T)
            slope_l = (regr.coef_).squeeze();       intercept_l = (regr.intercept_).squeeze()
            y_l = np.power(10,intercept_l)*np.power(f_idx,slope_l)
            
            regr = linear_model.LinearRegression()  
            regr.fit(np.log10(f_fit).reshape(-1,1), np.log10((psd_m_i[k,:][idx_f])[idx_f_fit]).T)
            slope_i = (regr.coef_).squeeze();       intercept_i = (regr.intercept_).squeeze()
            y_i = np.power(10,intercept_i)*np.power(f_idx,slope_i)
            
            np.save(fdn+'tracking/slope_l.npy',slope_l); np.save(fdn+'tracking/intercept_l.npy',intercept_l);
            np.save(fdn+'tracking/slope_i.npy',slope_i); np.save(fdn+'tracking/intercept_i.npy',intercept_i);
            
            ax.plot(f_idx, y_l, 'b-');  ax.plot(f_fit, y_l[idx_f_fit], 'b.');
            ax.plot(f_idx, y_i, 'r-');  ax.plot(f_fit, y_i[idx_f_fit], 'r.');
            
            psd_m_l_fit[k,:] = (psd_m_l[k,:][idx_f])/y_l;   y_m_l_fit[k,:] = y_l
            psd_m_i_fit[k,:] = (psd_m_i[k,:][idx_f])/y_i;   y_m_i_fit[k,:] = y_i
            
            y_m_i_fit[k,:] = y_i
            ax.set_ylim([5,60])
            
        np.save(rfdn+'tracking/psd_m_l_fit'+rats+'.npy',psd_m_l_fit);
        np.save(rfdn+'tracking/psd_m_i_fit'+rats+'.npy',psd_m_i_fit);
        np.save(rfdn+'tracking/y_m_l_fit'+rats+'.npy',y_m_l_fit);
        np.save(rfdn+'tracking/y_m_i_fit'+rats+'.npy',y_m_i_fit);
        plt.savefig('pic2/BBB_raw_'+rfdn[9:len(rfdn)-1]+rats+'.png')
        plt.clf()
        print(rfdn, n_k, np.sum(epn_clean))
        rr += 1
```
