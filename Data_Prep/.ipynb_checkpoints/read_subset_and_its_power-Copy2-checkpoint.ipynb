<h1> calculating each channel's correlation with speed</h1>

```python
segment = 0.7
fs_out = 400
for k in epn:
    
    fdn = rfdn+ k + '/'
    idx_viable = np.load(fdn+'NPY_'+str(fs_out)+'/idx_viable.npy')
    v_1d = np.load(fdn+'NPY_'+str(fs_out)+'/speed.npy');
    ts_f =  np.load(fdn+'NPY_'+str(fs_out)+'/ts_ds.npy');
    f = np.load(fdn + 'Beta_'+str(fs_out)+'/f.npy')
    nperseg = len(f)
    idx = np.arange(len(ts_f))
    idx_v = np.logical_not(np.isnan(v_1d))
    idx_t = np.logical_and(idx<=segment*len(ts_f),idx_viable)
    idx_vt = np.logical_and(idx_v,idx_t).squeeze()
    v = v_1d[idx_vt];
    idx_v_m = v>=3.5;
    idx_v_s = v<=1;
    
    pc_chan_m = np.empty([0, nperseg])
    pc_chan_s = np.empty([0, nperseg])
    for i in subset:
        chn = str(i//10)+str(i%10)
        Sxx_ch = np.load(fdn + 'Beta_'+str(fs_out)+'/ch.'+chn+'.S_beta.npy')[:,idx_vt]
        pc_chan_m_j = np.empty((0,))
        pc_chan_s_j = np.empty((0,))
        for j in np.arange(nperseg):
            pc_chan_m_j = np.append(pc_chan_m_j, pearsonr(Sxx_ch[j,idx_v_m],v[idx_v_m])[0])
            pc_chan_s_j = np.append(pc_chan_s_j, pearsonr(Sxx_ch[j,idx_v_s],v[idx_v_s])[0])
        pc_chan_m = np.vstack((pc_chan_m, pc_chan_m_j))
        pc_chan_s = np.vstack((pc_chan_s, pc_chan_s_j))
    np.save(fdn + 'Beta_'+str(fs_out)+'/pc_chan_m.npy',pc_chan_m)
    np.save(fdn + 'Beta_'+str(fs_out)+'/pc_chan_s.npy',pc_chan_s)
#     pc_chan_m = np.load(fdn + 'Beta_'+str(fs_out)+'/pc_chan_m.npy')
#     pc_chan_s = np.load(fdn + 'Beta_'+str(fs_out)+'/pc_chan_s.npy')
    
    pc_te_m = np.empty([0, nperseg])
    pc_te_s = np.empty([0, nperseg])
    for i in subset_t:
        idx_t = np.logical_and(subset<(i+1)*4, subset>=(i)*4)
        pc_m_t = np.mean(pc_chan_m[idx_t,:],axis=0)
        pc_s_t = np.mean(pc_chan_s[idx_t,:],axis=0)
        pc_te_m = np.vstack((pc_te_m, pc_m_t))
        pc_te_s = np.vstack((pc_te_s, pc_s_t))
    np.save(fdn + 'Beta_'+str(fs_out)+'/pc_te_m.npy',pc_te_m)
    np.save(fdn + 'Beta_'+str(fs_out)+'/pc_te_s.npy',pc_te_s)
    print(k)
```
